```r
## Use names better for visualization
add_spaces_dict <- c("B-0: CD24+CD27+CD11b+\n switched memory", 
                    "B-1: CD24++CD27+IgM+\n unswitched memory")
names(add_spaces_dict) <- c("B-0: CD24+CD27+CD11b+ switched memory", "B-1: CD24++CD27+IgM+ unswitched memory")
setdiff(names(add_spaces_dict), meta_data$cluster_name)
meta_data$cluster_name2 <- as.character(meta_data$cluster_name)
SA_micro_table$group2 <- as.character(SA_micro_table$group)
RA_micro_table$group2 <- as.character(RA_micro_table$group)
for (cell_group in names(add_spaces_dict)) {
    print(add_spaces_dict[cell_group])
    print(dim(meta_data[meta_data$cluster_name == cell_group,]))
    meta_data[meta_data$cluster_name == cell_group,]$cluster_name2 <- add_spaces_dict[cell_group]
    SA_micro_table[SA_micro_table$group == cell_group,]$group2 <- add_spaces_dict[cell_group]
    RA_micro_table[RA_micro_table$group == cell_group,]$group2 <- add_spaces_dict[cell_group]
    }
```

```r
# add cell scores to meta data
rownames(RA_cell_score) <- RA_cell_score$V1
rownames(SA_cell_score) <- SA_cell_score$V1
RA_cell_score <- RA_cell_score[meta_data$cell,]
SA_cell_score <- SA_cell_score[meta_data$cell,]
identical(RA_cell_score$V1, meta_data$cell)
identical(SA_cell_score$V1, meta_data$cell)
RA_meta_data <- cbind(meta_data, RA_cell_score)
SA_meta_data <- cbind(meta_data, SA_cell_score)
rm(meta_data)
gc()
```

```r
# list to add the ranked clusters within cell types
# base on RA for consistency
ranked_cell_clusters <- c()
for (cell_type in meta_colors1) {
    # get the ranked cells in that group
    data <- RA_micro_table[RA_micro_table$group2 %in% names(cell_type),]
    data <- data[order(data$assoc_mcz),]
    # now add the ranked cells to a list
    ranked_cell_clusters <- c(ranked_cell_clusters, data$group2)
    } 
ranked_cell_clusters
```

```r
# check no improper names
data_summary <- function(x) {
   m <- mean(x)
   ymin <- m-sd(x)
   ymax <- m+sd(x)
   return(c(y=m,ymin=ymin,ymax=ymax))
}
setdiff(RA_meta_data$cluster_name2, RA_micro_table$group2)
setdiff(RA_micro_table$group2, RA_meta_data$cluster_name2)
setdiff(RA_micro_table$group2, ranked_cell_clusters)
setdiff(ranked_cell_clusters, RA_micro_table$group2)
```

```r
# add the micro_table z-score to cluster info
RA_meta_data$Z_score <- rep(NA, nrow(RA_meta_data))
SA_meta_data$Z_score <- rep(NA, nrow(SA_meta_data))
for (cell_group in RA_micro_table$group2) {
    # get the z score to put there
    RA_meta_data[RA_meta_data$cluster_name2 == cell_group,]$Z_score <- RA_micro_table[RA_micro_table$group2 == cell_group,]$assoc_mcz
    SA_meta_data[SA_meta_data$cluster_name2 == cell_group,]$Z_score <- SA_micro_table[SA_micro_table$group2 == cell_group,]$assoc_mcz
    }
RA_meta_data[1:2,]
SA_meta_data[1:2,]
```
